```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
#importing all the necessary libraries
```

```python
#importing the dataset and representing as dt
dt = pd.read_csv('creditcard.csv')
```

```python
#retrieving the first 5 rows of dataset starting from 0 index.
dt.head()
```

```python
#used to analyze the numerical columns in dataframe.
dt.describe()
```

```python
#used to provide us with summary of data like index , count , memory usage , columns , non null values present.
dt.info()
```

```python
#To identify the number of transcations which are fraud and not fraud.
fraud = dt.loc[dt['Class'] == 1]
Notfraud = dt.loc[dt['Class'] == 0]
```

```python
#generating historgrams where each histogram represents distribution of numerical values across the dataframe.
dt.hist( bins = 30 , figsize = (15,15))  
```

```python
# X contains attributes of datset without the Class attribute ( input variable)
# Y contains only the Class attribute (output variable) 
x = dt.drop('Class', axis=1)
y = dt['Class']
```

```python
# Splitting the dataset into train and test sets to apply a model and find the accuracy. 
# x_train contains features for testing the x_test and same goes for y_train and y_test.
# test_size indicates that we have assumed 45% of data is going to be used for Testing and remaining 55% will be used for training.
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size = 0.45)
```

```python
#We use logistic regression model which can take up to 5000 items to find the optimal solution.
model = LogisticRegression(max_iter=5000) 
```

```python
model.fit(x_train, y_train)
```

```python
#using the y_pred varaible to store the predicted variables for further analysis
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
# To understand the accuracy using the prediction variable.
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
# Gives a summary of correct and incorrect predictions made by the model .
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(conf_matrix)
```

```python
class_report = classification_report(y_test, y_pred)
print("Classification Report:")
print(class_report)
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)
```

```python
# Visualizing the Confudion Matrix for better understanding 
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.xlabel('Predicted Class')
plt.ylabel('Actual Class')
plt.title('Confusion Matrix')
plt.show()
```
